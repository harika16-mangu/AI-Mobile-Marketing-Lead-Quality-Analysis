```python
def categorize_lead(status):
    if status == 'Closed':
        return 'Converted'
    elif status in ['EP Sent', 'EP Received', 'EP Confirmed']:
        return 'Potential Lead'
    elif status in ['Contacted - Doesn\'t Qualify', 'Contacted - Invalid Profile']:
        return 'No Lead'
    else:
        return 'Other'

lead['LeadCategory'] = lead['CallStatus'].apply(categorize_lead)
```

### High Quality Leads

```python
lead['Closed'] = lead['CallStatus'].str.lower() == 'closed'
```

```python
# Extract month as integer
lead['Month'] = lead['LeadCreated'].dt.month
```

```python
lead['IsConverted'] = lead['LeadCategory'] == 'Converted'
lead['IsPotential'] = lead['LeadCategory'] == 'Potential Lead'
```

```python
monthly_quality = lead.groupby('Month').agg(
    Total_Leads=('VendorLeadID', 'count'),
    Converted_Rate=('IsConverted', 'mean'),
    Potential_Rate=('IsPotential', 'mean')
).reset_index()

# Convert rates to %
monthly_quality['Converted_Rate'] *= 100
monthly_quality['Potential_Rate'] *= 100
```

```python
monthly_quality.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
fig, ax1 = plt.subplots(figsize=(6, 6))

# Bar: total leads
ax1.bar(monthly_quality['Month'], monthly_quality['Total_Leads'], color='lightgray', label='Total Leads')
ax1.set_xlabel('Month', color='black')
ax1.set_ylabel('Total Leads', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Line 1: Converted Rate
ax2 = ax1.twinx()
sns.lineplot(data=monthly_quality, x='Month', y='Converted_Rate', ax=ax2, color='dodgerblue', marker='o', label='Converted %')

# Line 2: Potential Rate
sns.lineplot(data=monthly_quality, x='Month', y='Potential_Rate', ax=ax2, color='green', marker='o', label='Potential Lead %')

ax2.set_ylabel('Conversion Rate (%)')
ax2.tick_params(axis='y')

# Title & formatting
plt.title('Monthly Trends: Total Leads, Converted %, and Potential %')
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Set seaborn style
sns.set_style("whitegrid")

# Create the figure and axes
fig, ax1 = plt.subplots(figsize=(6, 6))

# --- Bar chart for Closed Leads ---
ax1.bar(monthly_quality['Month'], monthly_quality['Total_Leads'], color='skyblue', label='Total Leads')
ax1.set_xlabel('Month')
ax1.set_ylabel('Total Leads', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.tick_params(axis='x', rotation=0)

# --- Line chart for Close Rate ---
ax2 = ax1.twinx()
sns.lineplot(x='Month', y='Close_Rate', data=monthly_quality, ax=ax2, color='darkred', marker='o', label='Close Rate (%)')
ax2.set_ylabel('Close Rate (%)', color='darkred')
ax2.tick_params(axis='y', labelcolor='darkred')

# Optional: Add a target reference line at 9.6%
ax2.axhline(y=9.6, color='gray', linestyle='--', linewidth=1, label='Target Close Rate')

# Title and legend
plt.title('Monthly Trend of Total Leads and Close Rate')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.tight_layout()
plt.show()
```

so from the above graph i observed the sessional trend supersingly in high conversion rates alternate months of april,june,aug(4,6,8) and low conversation rates in (5,7,9)
 1. Lead quality drops after high-performing months
Good leads may get exhausted, and weaker leads follow (especially if sources stay the same).

Advertisers may ramp up budget one month, then coast the next.

📌 Insight: You might have a "wave pattern" in traffic or ad quality.

🔹 2. Ad fatigue or creative fatigue
If the same creative is used over months, it may perform better initially, then degrade.

Perhaps new creatives launched in months 4, 6, and 8.

📌 Insight: Look at WidgetName by month — did high-performing creatives rotate in on 4, 6, 8?

🔹 3. Campaign structure or budget pacing
Media buyers may shift budgets aggressively every other month.

Or CPL targets change in alternate months — causing conversion prioritization.

📌 Insight: Look at Campaign, Partner, or PublisherCampaignName — see if anything changes structurally between months.

🔹 4. Seasonal behavior or market response
External factors may affect consumer readiness or advertiser follow-up speed.

For example, summer months or back-to-school planning can change priorities.

```python
#widget performance over months
lead['Month'] = pd.to_datetime(lead['LeadCreated']).dt.month
lead['IsClosed'] = lead['CallStatus'].str.lower() == 'closed'

widget_month_summary = lead.groupby(['Month', 'WidgetName']).agg(
    Total_Leads=('VendorLeadID', 'count'),
    Closed_Leads=('IsClosed', 'sum')
).reset_index()

widget_month_summary['Close_Rate'] = (widget_month_summary['Closed_Leads'] / widget_month_summary['Total_Leads']) * 100
```

```python
widget_month_summary
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
sns.barplot(
    data=widget_month_summary,
    x='Month',
    y='Total_Leads',
    hue='WidgetName',
    palette='Set2'
)
plt.title('Close Rate by WidgetName (Months 4 to 9)')
plt.xticks(rotation=0, ha='right')
plt.ylabel('Total Leads')
plt.tight_layout()
plt.legend(title='widget')
# 👇 Removes legend
plt.legend().set_visible(False)
plt.savefig("C:/Users/harik/Downloads/widget_leads_by_month.png", dpi=300, bbox_inches='tight')
plt.show()
```

widget perfoemance varies month to month some widgets appears to be in top and Consistently performing widgets which has the high potential to attract our leads are w-300250-DebtReduction1-1DC-CREDITSolutions,w-300250-DebtReduction1-1DC-white,w-300250-DebtReduction1-1DC,w-300250-DebtReduction1-1DC-Head3. these suggests that certain creatives may be more resilient across different traffic conditions — a valuable asset for campaign planning.

```python
# Sort and select top 10 widgets by total Closed Leads
top_widgets = widget_month_summary.groupby('WidgetName')['Closed_Leads'].sum().nlargest(5).index

# Filter your main DataFrame
top_widget_data = widget_month_summary[widget_month_summary['WidgetName'].isin(top_widgets)]
```

```python
plt.figure(figsize=(6, 6))
sns.barplot(
    data=top_widget_data,
    x='WidgetName',
    y='Closed_Leads',
    palette='Set2'
)
plt.title('Top 10 Widgets by Total Closed Leads')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Closed Leads')
plt.tight_layout()
plt.show()
```

the widget from which the leads are closing are w-300250-DebtReduction1-1DC-CREDITSolutions,w-300250-DebtReduction1-1DC-white,w-300250-DebtReduction1-1DC,w-300250-DebtReduction1-1DC-BlueMeter
